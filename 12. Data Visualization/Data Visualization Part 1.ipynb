```python
df['region_name'].str.contains('Los Angeles')
```

```python
df_Angeles = df[df['region_name'].str.contains('Los Angeles')]
df_Nevada = df[df['region_name'].str.contains('Nevada County, CA')]
df_San_Francisco = df[df['region_name'].str.contains('San Francisco')]
```

### concate

```python
df_target = pd.concat([df_Angeles,df_Nevada, df_San_Francisco])
```

```python
df_target[df_target['region_name']=='Nevada County, AR']
```

```python
df1 = df_target.groupby('region_name')['median_active_list_price'].sum()
```
